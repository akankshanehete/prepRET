```python
def show_all_channels(cropped_img):
  r,g,b=cv2.split(cropped_img)
  #clahe = cv2.createCLAHE(clipLimit=4, tileGridSize=(10, 10))
  #g = clahe.apply(g)
  plt.figure(figsize=(15, 8))
  plt.subplot(1, 4, 1)
  imshow(cropped_img)
  plt.title('Original')
  plt.subplot(1, 4, 2)
  io.imshow(b)
  plt.title('Blue Channel')
  plt.subplot(1, 4, 3)
  plt.title('Green Channel')
  io.imshow(g)
  plt.subplot(1, 4, 4)
  plt.title('Red Channel')
  io.imshow(r )

def imshow(image):
  image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
  io.imshow(image_rgb)
```
